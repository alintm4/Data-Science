```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

days = 100  
initial_price = 100  
daily_returns = np.random.normal(0, 0.02, days)
stock_prices = initial_price * (1 + daily_returns).cumprod()

plt.plot(stock_prices)
plt.title("Simulated random Stock")
plt.xlabel("Days")
plt.ylabel("Price")
plt.show()

```

```python
mean_price = np.mean(stock_prices)
std_dev = np.std(stock_prices)
total_return = (stock_prices[-1] - stock_prices[0]) / stock_prices[0] * 100

print(f"Mean Stock Price: {mean_price:.2f}")
print(f"Stock Price Standard Deviation: {std_dev:.2f}")
print(f"Total Return: {total_return:.2f}%")
```
