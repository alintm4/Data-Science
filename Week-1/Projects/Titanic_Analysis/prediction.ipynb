```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('eda_titanic.csv')
```

```python
class_survival = data.groupby('Pclass')['Survived'].mean()
print("\nSurvival rate by Pclass:")
print(class_survival)

data['AgeGroup'] = pd.cut(data['Age'], bins=[0, 12, 18, 30, 50, 80], 
                          labels=['Child', 'Teen', 'Young Adult', 'Adult', 'Senior'])
age_survival = data.groupby('AgeGroup')['Survived'].mean()
print("\nSurvival rate by Age group:")
print(age_survival)

gender_survival = data.groupby('Sex')['Survived'].mean()
print("\nSurvival rate by Sex:")
print(gender_survival)
```

```python
sns.barplot(x=class_survival.index, y=class_survival.values)
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()

sns.barplot(x=age_survival.index, y=age_survival.values)
plt.title('Survival Rate by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Survival Rate')
plt.show()

sns.barplot(x=gender_survival.index, y=gender_survival.values)
plt.title('Survival Rate by Sex')
plt.xlabel('Sex')
plt.ylabel('Survival Rate')
plt.show()
```
