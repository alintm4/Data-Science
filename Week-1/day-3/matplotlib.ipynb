```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6]
y=[2,6,3,4,5,0]

plt.plot(x,y,label='random',color='red',linewidth='4',marker='.',markersize=20,linestyle='--')
plt.xlabel("Using matplotlib x-axis")
plt.ylabel("Using matplotlib y-axis")
plt.title("A basic Line graph")
x2=np.arange(0,6,1)

plt.plot(x2[:4],x2[:4]**2,label='x^2',linestyle='--')
plt.xticks([0,1,2,3,4,5,6])
plt.yticks([0,1,2,3,4,5,6])


plt.legend()

plt.show()
```

# bar graph

```python
labels=['A','B','C']
values=[1,4,2]

bars=plt.bar(labels,values)
bars[0].set_hatch('o')
plt.figure(figsize=(6,4))
plt.show()
```

## Stock Prices of Netflix visualizing

```python
data=pd.read_csv("NFLX.csv")
display(data.head())
```

```python
plt.plot(data.Date, data.High)
plt.show()
plt.plot(data.High, data.Close)
plt.show()
```

```python
plt.plot(data.Date, data.Close, label='Close Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Stock Closing Price Over Time')
plt.legend()
plt.show()
```

```python
plt.bar(data.Date, data.Volume)
plt.xlabel('Date')
plt.ylabel('Volume')
plt.title('Trading Volume Over Time')
plt.show()
```

```python
plt.fill_between(data['Date'], data['High'], data['Low'])
plt.title('High-Low Price Range Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

```python
data['Daily Change'] = data['Close'] - data['Open']
plt.bar(data['Date'], data['Daily Change'], color='green')
plt.title('Daily Price Changes')
plt.xlabel('Date')
plt.ylabel('Change')
plt.show()
```

```python
plt.plot(data['Date'], data['Open'], label='Open Price')
plt.plot(data['Date'], data['Close'], label='Close Price')
plt.plot(data['Date'], data['High'], label='High Price')
plt.plot(data['Date'], data['Low'], label='Low Price')
plt.title('Stock Prices (Open, Close, High, Low)')
plt.legend()
plt.show()
```
