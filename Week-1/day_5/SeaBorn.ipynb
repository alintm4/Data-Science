```python
import pandas as pd
import numpy as  np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sea=sns.load_dataset("car_crashes")
sea
```

```python
sns.displot(sea['not_distracted'],bins=25)
```

```python
sns.jointplot(x='speeding',y='alcohol',data=sea,kind="kde")
#sns.jointplot(x='speeding',y='alcohol',data=sea,kind="hex")
```

```python
sns.pairplot(sea)
```

```python
tips_df=sns.load_dataset("tips")

sns.pairplot(tips_df,hue="sex")
```

```python
sns.barplot(x='sex',y='total_bill',data=tips_df)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips_df,hue="sex") #for outliers
```

```python
#heatmaps
numeric_sea = sea.select_dtypes(include=['number'])

sea_mx = numeric_sea.corr()

plt.figure(figsize=(8, 6))
sns.set_context('paper', font_scale=1.4)
sns.heatmap(sea_mx, annot=True, cmap='coolwarm')
plt.show()
```
